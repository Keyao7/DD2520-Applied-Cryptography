# Fixed XOR

```python
def main(str1, str2):
    list_hex_1 = []
    list_hex_2 = []
    bin_str_1 = ""
    bin_str_2 = ""
    XOR_bin = ""
    XOR_str = ""

    # Split input by groups of two
    while len(str1) != 0:
        list_hex_1.append(str1[:2])
        list_hex_2.append(str2[:2])
        str1 = str1[2:]
        str2 = str2[2:]

    # Converts hex to binary and unifies it into eight binary digits.
    for i in range(len(list_hex_1)):
        tmp_bin_1 = bin(int(list_hex_1[i], 16))[2:]
        tmp_bin_2 = bin(int(list_hex_2[i], 16))[2:]
        for j in range(8 - len(tmp_bin_1)):
            tmp_bin_1 = "0" + tmp_bin_1
        for j in range(8 - len(tmp_bin_2)):
            tmp_bin_2 = "0" + tmp_bin_2
        bin_str_1 += tmp_bin_1
        bin_str_2 += tmp_bin_2

    # Compare each binary digit of the two groups and 
    # return 0 if they are the same, or 1 if they are not.
    for i in range(len(bin_str_1)):
        if bin_str_1[i] == bin_str_2[i]:
            XOR_bin += "0"
        else:
            XOR_bin += "1"
    
    # Convert binary to hex
    while len(XOR_bin) != 0:
        XOR_str += hex(int(XOR_bin[:8],2))[2:]
        XOR_bin = XOR_bin[8:]
    return XOR_str
    
```

```python
input_hex = "1c0111001f010100061a024b53535009181c"
input_hex_2 = "686974207468652062756c6c277320657965"
main(input_hex, input_hex_2)
```
