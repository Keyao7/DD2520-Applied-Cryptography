```python
def scoring(text):
    score = 0
    # a - z: 97 - 112
    # A - Z: 65 - 90
    # 0 - 9: 49 - 57
    # space: 32
    for char in text:
        if ord(char) == 32:
            score += 3
        if 97 <= ord(char)<= 112 or 65 <= ord(char) <= 90 :
            score += 2
        if 49 <= ord(char) <= 57:
            score += 1
    return score
```
