```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
```

```python
key_space = '0123456789ABCDEFGHIJKLMNOPQRSTUVWXYZ'
len(key_space)
```

```python
cipher = ''
with open("Ciphertext_1.txt", "r") as f:
    for line in f.readlines():
        line = line.strip('\n')
        cipher += line
    print(type(cipher))
    print(len(cipher))
cipher_new = cipher.replace("_", " ")
```

```python
cipher_lines = []
```

```python
substring = ""
for i in range(len(cipher)):
    if cipher_new[i] != "#":
        substring += cipher_new[i]
    else:
        cipher_lines.append(substring)
        substring = ""
```

```python
cipher_lines
```

Frequency Analysis

```python
char_dic = {}
for i in range(len(key_space)):
    key_char = key_space[i]
    char_dic[key_char] = 0
for i in range(len(cipher_new)):
    char = cipher_new[i]
    if char != "#" and char != " ":
        char_dic[char] += 1
char_dic_sorted = sorted(char_dic.items(),key = lambda x:x[1],reverse = True)
char_dic_sorted

x_values = []
y_values = []
for i in range(len(char_dic_sorted)):
    y_values.append(char_dic_sorted[i][1])
    x_values.append(char_dic_sorted[i][0])

plt.bar(x_values, y_values)

plt.show()
```

```python
two_char_dic = {}

for i in range(len(cipher)-2):
    if cipher[i-1] == '_' and cipher[i] != '_' and cipher[i+1] != '_' and cipher[i+2] == '_':
        tmp_char = cipher[i] + cipher[i+1]
        if tmp_char not in two_char_dic.keys():
            two_char_dic[tmp_char] = 1
        else:
            two_char_dic[tmp_char] += 1

two_char_dic_sorted = sorted(two_char_dic.items(),key = lambda x:x[1],reverse = True)

x_values = []
y_values = []
for i in range(len(two_char_dic_sorted)):
    y_values.append(two_char_dic_sorted[i][1])
    x_values.append(two_char_dic_sorted[i][0])
plt.figure(figsize=(15, 8))
plt.bar(x_values, y_values)

plt.show()
```

```python
three_char_dic = {}

for i in range(1, len(cipher)-2):
    if cipher[i-1] == '_' and cipher[i] != '_' and cipher[i+1] != '_' and cipher[i+2] != '_' and cipher[i+3] == '_':
        tmp_char = cipher[i] + cipher[i+1] + cipher[i+2]
        if tmp_char not in three_char_dic.keys():
            three_char_dic[tmp_char] = 1
        else:
            three_char_dic[tmp_char] += 1
three_char_dic_sorted = sorted(three_char_dic.items(),key = lambda x:x[1],reverse = True)


x_values = []
y_values = []
for i in range(13):
    y_values.append(three_char_dic_sorted[i][1])
    x_values.append(three_char_dic_sorted[i][0])
plt.figure(figsize=(15, 8))
plt.bar(x_values, y_values)

plt.show()
```

```python
four_char_dic = {}

for i in range(1, len(cipher)-2):
    if cipher[i-1] == '_' and cipher[i] != '_' and cipher[i+1] != '_' and cipher[i+2] != '_' and cipher[i+3] != '_' and cipher[i+4] == '_':
        tmp_char = cipher[i] + cipher[i+1] + cipher[i+2] + cipher[i+3]
        if tmp_char not in four_char_dic.keys():
            four_char_dic[tmp_char] = 1
        else:
            four_char_dic[tmp_char] += 1
four_char_dic = sorted(four_char_dic.items(),key = lambda x:x[1],reverse = True)


x_values = []
y_values = []
for i in range(13):
    y_values.append(four_char_dic[i][1])
    x_values.append(four_char_dic[i][0])
plt.figure(figsize=(15, 8))
plt.bar(x_values, y_values)

plt.show()
```
